```python
import os
import glob
import configparser
import psycopg2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
config = configparser.ConfigParser()
config.read('dwh.cfg')

conn = psycopg2.connect("host={} dbname={} user={} password={} port={}".format(*config['CLUSTER'].values()))
cur = conn.cursor()
```

```python
def graphTop(query,top,title,conn,others):
    try: 
        cur.execute(query)
    except psycopg2.Error as e: 
        print("Error: select *")
        print (e)
        conn.close()
    rows = cur.fetchall()
    data = [x[0] for x in rows]
    topData = []
    i = 0
    sums = 0
    for dat in data:
        if i<top:
            topData.append(dat)
        elif others:
            sums+=dat
        i+=1
    labels=[x[1] for x in rows]
    labelsTop = labels[:top]
    if others:
        labelsTop.append('Others')
        topData.append(sums)
    fig, ax = plt.subplots(figsize=(15, 10), subplot_kw=dict(aspect="equal"))
    def func(pct, allvals):
        absolute = int(pct/100.*np.sum(allvals))
        return "{:.1f}%\n({:d})".format(pct, absolute)
    wedges, texts, autotexts = ax.pie(topData, autopct=lambda pct: func(pct, topData),
                                      textprops=dict(color="w"),counterclock=False, startangle=90)

    ax.legend(wedges, labelsTop,
              title=title,
              loc="center left",
              bbox_to_anchor=(1, 0, 0.5, 1))

    plt.setp(autotexts, size=14, weight="bold")

    ax.set_title(title)
    plt.savefig('./images/analytics/'+title+'.png')
    plt.show()
```

```python
graphTop('SELECT count(start_time),location from songplays group by location order by count(start_time) desc',
        5,'Top 5 Location',conn,others=True)
```

```python
graphTop('Select count(start_time),user_agent from songplays group by user_agent order by count(start_time) desc',
        5,'Top 5 User Agent',conn,others=True)
```

```python
try: 
    cur.execute("""
    select count(start_time) as counter, users.user_id, users.first_name, users.gender
    from songplays 
    join users 
    on users.user_id = songplays.user_id 
    group by users.user_id, users.first_name,users.gender
    order by counter DESC;
    """)
    rows = cur.fetchall()
except psycopg2.Error as e: 
    print("Error: select *")
    print (e)
    conn.close()
```

```python
dfpeople = pd.DataFrame(rows,columns=['Count','User Id','Name','Gender'])
dfpeople.to_csv("./images/analytics/people.csv")
```
